# 03 — Preprocessing & Cleaning

**Goal:** Transform the raw text file into a clean, continuous time-series dataset ready for feature engineering.
Based on our EDA, we need to: merge datetime, handle block-missing values (forward-fill), and remove redundant columns.

## 1. Imports & Setup

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
print('Setup complete.')
```

## 2. Load & Clean Data

```python
RAW_PATH = '../data/raw/household_power_consumption.txt'
CLEANED_PATH = '../data/processed/cleaned.csv'

# Load data, treating '?' as NaN
df = pd.read_csv(RAW_PATH, sep=';', na_values='?', low_memory=False)

# Create proper DatetimeIndex
df['datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d/%m/%Y %H:%M:%S')
df.set_index('datetime', inplace=True)
df.drop(columns=['Date', 'Time'], inplace=True)

# Handle missing values (Forward Fill for block outages)
df.ffill(inplace=True)

# Drop Global_intensity due to near-perfect multicollinearity with Global_active_power
df.drop(columns=['Global_intensity'], inplace=True)

print(f'Cleaned data shape: {df.shape}')
df.head(3)
```

## 3. Save Cleaned Data

```python
# Save to processed folder so notebook 04 can pick it up
df.to_csv(CLEANED_PATH)
print(f'Saved cleaned data to {CLEANED_PATH}')
```
